```python
df_poids = df_poids[["Manufacturer","Weight"]]
```

```python
df_poids[['Weight_Min', 'Weight_Max']] = df_poids['Weight'].str.split(' - ', expand=True)

# Convert the new columns to numeric types (integers or floats)
df_poids['Weight_Min'] = pd.to_numeric(df_poids['Weight_Min'], errors='coerce')
df_poids['Weight_Max'] = pd.to_numeric(df_poids['Weight_Max'], errors='coerce')
```

```python
df_poids
```

```python
df_poids = df_poids[["Manufacturer","Weight_Min","Weight_Max"]]
```

```python
df_poids
```

```python
cols_to_convert = ["Weight_Min","Weight_Max"]
for col in cols_to_convert:
    df_poids[col] = pd.to_numeric(df_poids[col], errors="coerce")


df_poids[cols_to_convert] = df_poids[cols_to_convert].astype("Int64")
```

```python
df_poids
```

```python
df_poids["Manufacturer"].unique()
```

```python
df_poids = df_poids.dropna(subset=["Manufacturer", "Weight_Min", "Weight_Max"])
df_poids
```
